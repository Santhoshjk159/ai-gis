# Exploration Notebook
Quick exploratory analysis for endangered medicinal plant location data and GIS-ready transformations.

```python
import pandas as pd
from pathlib import Path

root = Path('..')
df = pd.read_csv(root / 'data' / 'processed' / 'sample_locations.csv')
df.head()
```

```python
summary = df.groupby('species').agg(
    count=('species', 'count'),
    avg_suitability=('environmental_suitability', 'mean'),
    avg_elevation=('elevation_m', 'mean')
).sort_values('avg_suitability', ascending=False)
summary
```
